```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
compare_list = []
compare_list.append(list(range(0,11)))
```

```python
with open("epoch_loss_panetal-full.txt") as file:
    lines = file.readlines()
    lines = [line.rstrip() for line in lines]
    
a_li = []
for i in range(len(lines)):
    epoch = lines[i].split("/")[0].split(" ")[1]
    miniepoch = i
    trloss = lines[i].split(": ")[1]
    a_li.append([epoch, miniepoch, trloss])
    
df = pd.DataFrame(a_li, columns=["epoch", "miniepoch", "trloss"])
df["trloss"] = pd.to_numeric(df["trloss"])

df.plot(x="miniepoch", y="trloss")
plt.show()

interval = df.groupby(["epoch"], as_index=False).last()["miniepoch"].values[0]

counter = 0
i_li = []
for i in range(0, len(df), interval):
    #print(df["miniepoch"][i], df["trloss"][i])
    i_li.append([counter, df["trloss"][i]])
    counter += 1
    
df1 = pd.DataFrame(i_li, columns=["epoch", "trloss"])
df1.plot(x="epoch", y="trloss")
plt.show()

compare_list.append(list(df1["trloss"]))
```
